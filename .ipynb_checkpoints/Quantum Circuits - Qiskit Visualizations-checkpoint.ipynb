```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
circ = QuantumCircuit(2,2)
circ.h(0)
circ.cx(0,1) #Bell state generated: (1/sqrt(2))(|00> + |11>)

meas = QuantumCircuit(2,2)
meas.measure([0,1],[0,1]) #Generate a measure circuit which we will compose with circ

backend = BasicAer.get_backend('qasm_simulator')
qc = circ.compose(meas)
qc_transpiled = transpile(qc,backend)
job = backend.run(qc_transpiled, shots = 1000)
result = job.result()
counts = result.get_counts(circ)
print(counts)
```

```python
plot_histogram(counts)
```

```python
"""
Customizing histograms
"""
second_result = backend.run(qc_transpiled, shots = 1000).result()
second_counts = second_result.get_counts(circ)
legend = ['First execution','Second execution']
plot_histogram([counts,second_counts], legend = legend)
```

```python
plot = plot_histogram([counts,second_counts],legend=legend, sort = 'desc',figsize=(15,12),color = ['orange','black'], bar_labels=True)
```

```python
plot
```

```python
plot.savefig('my_plot_Qiskit_Visualizations.png')
```

```python
"""
Plot state
- Several functions for generating different types of visualization of a quantum state
    - 'plt_state_city': plots the real/imaginary parts of density matrix
    - 'plot_state_qsphere': amplitude and phase of statevector are plotted as a spherical ball. For mixed states, multiple sphers are plotted
    - 'plot_state_paulivec': representation of density matrix using pauli matrices
    - 'plot_state_hinton': same as the city,  but where the size of the element represents the value of the matrix element
    - 'plot_bloch_multivector': The projection of the quantum state onto the single qubit space and plotting on a bloch sphere
"""
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.visualization import plot_state_paulivec, plot_state_hinton
from qiskit.visualization import plot_state_qsphere
```

```python
state_backend = BasicAer.get_backend('statevector_simulator') #Need this backend to plot states
result2 = state_backend.run(transpile(circ, state_backend)).result()
psi = result2.get_statevector(circ)
```

```python
plot_state_city(psi)
```

```python
plot_state_hinton(psi)
```

```python
plot_state_qsphere(psi)
```

```python
plot_state_paulivec(psi)
```

```python
plot_bloch_multivector(psi) #A bell state doess not have any single-qubit information
```

```python
"""
Options when using state plotting functions
- 'plot_state_city'
    - title(str): a string that represents the plot title
    - figsize(tuple): figure size in inches (width,height)
    - color(list): a list of len=2 giving colors for real and imaginary components of matrix elements
- 'plot_state_hinton'
    - title(str): a string that represents the plot title
    - figsize(tuple): figure size in inches (width,height)
- 'plot_state_paulivec'
    - title(str): a string that represents the plot title
    - figsize(tuple): figure size in inches (width,height)
    - color(list or str): color of the expectation value bars
- 'plot_state_qsphere'
    - figsize(tuple): figure size in inches (width,height)
- 'plot_bloch_multivector'
    - title(str): a string that represents the plot title
    - figsize(tuple): figure size in inches (width,height)
"""
plot_state_city(psi, title="My City", color=['blue','orange'])
```

```python
plot_state_hinton(psi, title="My Hinton")
```

```python
plot_state_paulivec(psi,title="My Paulivec", color = ['purple','orange','green','blue'])
```

```python
plot_bloch_multivector(psi,title="My Bloch Spheres")
```

```python
"""
Plot Bloch Vector
-Options for plot_bloch_vector
    - title(str): a string that represents the plot title
    - figsize(tuple): figure size in inches (width, height)
"""
from qiskit.visualization import plot_bloch_vector
plot_bloch_vector([0,1,0])
```
