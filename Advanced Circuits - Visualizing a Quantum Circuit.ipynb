```python
from qiskit import *
```

```python
"""
Drawing a Quantum Circuit
"""
circuit = QuantumCircuit(3,3)
circuit.x(1)
circuit.h(range(3))
circuit.cx(0,1)
circuit.measure(range(3),range(3))
print(circuit)
```

```python
circuit.draw()
```

```python
"""
Alternative Renderers for Circuits
"""
circuit.draw('mpl')
```

```python
"""
Controlling output from circuit.draw()
    - 'text': returns TextDrawer object
    - 'mpl' returns matplotlib.Figure object
    - 'latex' returns PIL>Image object
Customizing the output
"""

q_a = QuantumRegister(3, name = 'qa')
q_b = QuantumRegister(5, name = 'qb')
c_a = ClassicalRegister(3)
c_b = ClassicalRegister(5)

circuit = QuantumCircuit(q_a,q_b,c_a,c_b)

circuit.x(q_a[1])
circuit.x(q_b[1])
circuit.x(q_b[2])
circuit.x(q_b[4])
circuit.barrier()
circuit.h(q_a)
circuit.barrier(q_a)
circuit.h(q_b)
circuit.cswap(q_b[0],q_b[1],q_b[2])
circuit.cswap(q_b[2],q_b[3],q_b[4])
circuit.cswap(q_b[3],q_b[4],q_b[0])
circuit.barrier(q_b)
circuit.measure(q_a,c_a)
circuit.measure(q_b,c_b)
circuit.draw(output = 'mpl')
```

```python
circuit.draw(output = 'mpl', reverse_bits = True) #Reversing bit order
```

```python
circuit.draw('mpl', plot_barriers = False) #Draw circuit without barriers
```

```python
circuit.draw('mpl',plot_barriers=False,reverse_bits=True)
```

```python
#Backend-specific customizations
style = {'backgroundcolor':'lightgreen'}
circuit.draw(output = 'mpl',style = style)
```

```python
circuit.draw(output='mpl',scale=0.5)
```

```python
"""
circuit_drawer()
    - Alternative to using a method of a QuantumCircuit object
    - Part of the public stable interface from qiskit.tools.visualization
    - Behaves identically to QuantumCircuit.draw(), but takes the circuit object as an argument
"""
from qiskit.tools.visualization import circuit_drawer
circuit_drawer(circuit,output='mpl',plot_barriers=False, style = style)
```
