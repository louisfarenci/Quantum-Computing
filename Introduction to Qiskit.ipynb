```python
"""
STEP 1: Import necessary packages.
"""
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, circuit_drawer
from matplotlib import *
from pylatexenc import *
```

```python
"""
STEP 2: Initialize variables
QuantumCircuit(qubits,classical_bits)
    -> Each parameter can be a QuantumRegister/ClassicalRegister object (respectively), an int, or a list of ints
"""
circuit = QuantumCircuit(2,2)
```

```python
"""
STEP 3: Add gates
"""
circuit.h(0) #Hadamard on qubit 0 (q0)
circuit.cx(0,1) #CNOT gate, with q0 as the control and q1 as the target
circuit.measure([0,1],[0,1]) #Measures q0 and q1, states are stored in c0 and c1 respectively
```

```python
"""
STEP 4: Visualize the circuit
"""
circuit.draw("mpl")
```

```python
"""
STEP 5: Simulate the experiment
"""
simulator = AerSimulator() #Simulator object needed for simulating the system
compiled_circuit = transpile(circuit,simulator) #Passes quantum circuit over to the simulator
job = simulator.run(compiled_circuit, shots = '1000') #Runs the identical circuit a number of times equal to shots
result = job.result() #retrives result
counts = result.get_counts(circuit)
print("Total counts for 00 and 11 are ",counts)
```

```python
"""
STEP 6: Visualize the results
"""
plot_histogram(counts)
```
